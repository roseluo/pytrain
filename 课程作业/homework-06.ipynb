### 1. 编程爬虫程序，尝试以统计学（https://baike.baidu.com/item/%E7%BB%9F%E8%AE%A1%E5%AD%A6/1175）为起始页面，抓取100条百度百科词条页面，结果截图

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import datetime
import random
import re


html = urlopen("https://baike.baidu.com/item/%E7%BB%9F%E8%AE%A1%E5%AD%A6/1175")

bsObj = BeautifulSoup(html,'lxml')
#print(bsObj.prettify())


links=[]
for link in bsObj.find("div", "main-content").find_all("a", href = re.compile("^(/item/)((?!:).)*$")):
    if 'href' in link.attrs:
        links.append(link)
        print(link.attrs['href'])

```
